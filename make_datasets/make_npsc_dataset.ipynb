# Make NPSC huggingface dataset for ASR training

```python
import pandas as pd
from datasets import load_dataset, Audio
from pathlib import Path
```

```python
cols = [
    "speaker_id",
    "gender",
    "utterance_id",
    "language",
    "raw_text",
    "full_audio_file",
    "original_data_split",
    "region",
    "duration",
    "start",
    "end",
    "utterance_audio_file",
    "standardized_text",
]
```

```python
npsc_data_files_raw = {"train": "raw_datasets/npsc_20230105_train.csv", "test": "raw_datasets/npsc_20230105_test.csv", "validation": "raw_datasets/npsc_20230105_eval.csv"}
```

```python
dataset_name = "npsc"
for k, v in npsc_data_files_raw.items():
    for lang in ["nb", "nn"]:
        df = pd.read_csv(v, names=cols).dropna(subset=["utterance_audio_file", "standardized_text"]).query(f"language == '{lang}-NO' & duration > 1 & duration < 15 & utterance_id != 'npsc_62309'") # possibly add other cleanup methods here
        outpath = dataset_name + "_" + k + "_" + lang + ".csv"
        df.to_csv(outpath, index=False)
        print(outpath)
```

```python
npsc_nb_data_files = {"train": "npsc_train_nb.csv", "test": "npsc_test_nb.csv", "validation": "npsc_validation_nb.csv"}
npsc_nn_data_files = {"train": "npsc_train_nn.csv", "test": "npsc_test_nn.csv", "validation": "npsc_validation_nn.csv"}
```

```python
dataset_nb_npsc = load_dataset("csv", data_files=npsc_nb_data_files, cache_dir="/media/pers/storage/storage/huggingface_cache/").cast_column("utterance_audio_file", Audio())
dataset_nn_npsc = load_dataset("csv", data_files=npsc_nn_data_files, cache_dir="/media/pers/storage/storage/huggingface_cache/").cast_column("utterance_audio_file", Audio())
```

```python
import IPython.display as ipd
import numpy as np
import random

rand_int = random.randint(0, len(dataset_nb_npsc["train"])-1)

print(dataset_nb_npsc["train"][rand_int]["standardized_text"])
ipd.Audio(data=dataset_nb_npsc["train"][rand_int]["utterance_audio_file"]["array"], autoplay=True, rate=16000)
```
