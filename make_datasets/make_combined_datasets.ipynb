# Make combined huggingface dataset for ASR training

```python
import pandas as pd
from datasets import load_dataset, Audio
from pathlib import Path
```

```python
data_files = {"train": "combined_datasets/npsc_rundkast_sample_nb_train.csv", "test": "combined_datasets/npsc_rundkast_sample_nb_test.csv", "validation": "combined_datasets/npsc_rundkast_sample_nb_eval.csv"}
```

```python
dataset = load_dataset("csv", data_files=data_files, cache_dir="/path/to/cache_dir").cast_column("utterance_audio_file", Audio())
```

```python
import IPython.display as ipd
import numpy as np
import random

rand_int = random.randint(0, len(dataset["train"])-1)

print(dataset["train"][rand_int]["standardized_text"])
ipd.Audio(data=dataset["train"][rand_int]["utterance_audio_file"]["array"], autoplay=True, rate=16000)
```

```python
dataset.push_to_hub(repo_id="scribe-project/NPSCRundkast_nb_sample", max_shard_size="200GB", embed_external_files=True)
```
